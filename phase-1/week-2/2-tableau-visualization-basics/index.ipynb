# Tableau - Useful Visualizations and the Show Me Pane

## Introduction
Data visualizations are not one-size fits all so it is important to select a visualization that works for your data and the story that you want to tell with that data. Tableau can generate a diverse collection of __vizzes__ (Tableau's special word for visualizations) that you can customize for your use case. In this lesson, we will discuss __Measures__ and __Dimensions__ and how you should choose them to build a viz. Then, we discuss Show Me pane and highlight a few of the most essential visualizations - the Bar Chart, the Scatter Plot, and the Line Plot. 

## Objectives
You will be able to:
* Identify the dimensions and measures needed to generate a viz
* Apply a visualization to data using the Show Me Pane

## Measures and Dimensions
If we return to the Data Pane, you might notice that the names of the columns have two distinct colors, <font color='green'>__green__</font> and <font color='blue'>__blue__</font>. These colors help to distinguish between __Measures__ and __Dimensions__. These will be the building blocks of your visualizations.

<h3><font color='green'><b>Measures</b></font></h3>
<b>Measures are continuous data points.</b> They contain the quantitative values that can be measured such as sales, height, and age. <br>
<li>These values can be aggregated. When you add measure to your workspace Tableau automatically applies a default aggregation, which is usually sum.</li>
<li>If you look just before the name of each measure, you will see a small icon. These icons represent data type and are automatically generated by Tableau.</li>
<li>If the data type is not accurate or you just want to customize it, you can do so by hovering over the measure, and then clicking the arrow to open the options menu.
</li>
<h3><font color='blue'><b>Dimensions</b></font></h3>
<b>Dimensions are categorical data points.</b> They contain qualitative values like month, state, and date.<br>

<li>When added to the workspace, dimensions will impact he level of detail in the view.</li>
<li>If you look just before the name of each dimension, you will see a small icon. These icons represent data type and are automatically generated by Tableau and can be modified.</li>
